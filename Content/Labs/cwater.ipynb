```{solution}
Eerst moet de maatbeker gewogen worden, daarna wordt  300 mL water toegevoegd aan de maatbeker en opnieuw gewogen en wordt het verschil genoteerd. Vervolgens wordt de temperatuursensor in het water gezet, waarna het elektrische verwamingselement, dat uit staat, volgt. Vervolgens wordt een roervlo toegevoegd en wordt de machine voor de roervlo aangezet, zodat de temperatuur door het water heen homogeen is. Dan wordt tegelijk het elektrisch verwamingselement aangezet en de stopwatch gestart. Om zo veel mogelijk verschil in warmte te meten, wordt het verwamingselement op 40 W gezet.  Vervolgens wordt om de 30 seconde de temperatuur van het water genoteerd. Doe dit voor 10 minuten lang.

Aanpassingen:
Aangezien het vermogen van te voren gelijk op 40 W kan worden gezet (aangezien het verwamingselement uit moet) zullen de eerste 10 seconden het vermogene van het verwamingselement te laag zijn wat resulteert in een gereduceerde energie toevoeging. Om dit te verbeteren zal het verwarmings element bijvoorbeeld eerst in ander water worden geplaatst, zodat de bron zijn spanning en stroom ingesteld kunnen worden voor een vermogen van 40 W (als deze dan in het aan wordt gezet bij de proef, zal het gelijk een vermogen van 40 W hebben).

Er wordt niks gezegd over waar exact de temperatuursensor in het water moet. als het namelijk dicht bij het verwarmings element ligt in het water, zal het een hogere dan gemiddelde temperatuur meten. 

Over de stand van de roervlo wordt niks gezegd. Is deze te laag, zal de temperatuur niet homogeen zijn door het water heen. Is deze te hoog, zal het glas kapot gaan.

Een tijd is indicator voor hoelang het door moet gaan kan beter worden aangepast naar een eind temperatuur als stopindicator. Met een eind temperatuur als stopindicator heb je gegarandeerd genoeg meetpunten. tijd is ook goed, aangezien 10 minuten genoeg meetpunten heeft, maar temperatuur kan duidelijk nog meer groeien na 10 minuten, dus zal de tijd wel langer moeten worden.
```
